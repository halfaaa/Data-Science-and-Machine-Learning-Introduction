# Manipulating Data

```python
# lowering the strings 
car_sales["Make"].str.lower()
```

```python
car_sales_missing=pd.read_csv("009 car-sales-missing-data.csv")
car_sales_missing
```

```python
# Filling NaN values with mean value
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
# inplacing values
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(),
                                    inplace=True)
car_sales_missing
```

```python
# Removing rows including NaN values
car_sales_missing_dropped=car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
# column from series
seats_column =pd.Series([5,5,5,5,5])

# New column called seats
car_sales["Seats"]=seats_column
car_sales
```

```python
# fill NaN values
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

```python
# Column from Python List
fuel_economy = [7.5,9.2,5.0,9.6,8.7,4.7,7.6,8.6,3.1,4.2]
car_sales["Fuel per 100KM"]=fuel_economy
car_sales
```

```python
# Create another New Column
car_sales["Total fuel used (L)"]=car_sales["Odometer (KM)"]/100*car_sales["Fuel per 100KM"]
car_sales
```

```python
# Create a Column from a single value
car_sales["Number of wheels"]=4
car_sales
```

```python
# Create a boolean column
car_sales["Passed road saftey"]=True
car_sales.dtypes
```

```python
car_sales
```

```python
# shuffling indexes
car_sales_shuffled=car_sales.sample(frac=1)
car_sales_shuffled
```

```python
# Only select 20% of data
car_sales_shuffled.sample(frac=0.2)
```

```python
# reset shuffled dataframe
car_sales_shuffled.reset_index(drop=True,inplace=True)
car_sales_shuffled
```

```python
# Convert Odometer KM to Mile
car_sales["Odometer (KM)"]=car_sales["Odometer (KM)"].apply(lambda x:x/1.6)
car_sales
```
